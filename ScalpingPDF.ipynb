#Notebook is for extracting all text from the first page of each pdf documents in a folder and write them into individual text files.

Please create a folder named "pdf_doc" in your current working folder (where this notebook file resides).

Warning: Do not expand the columns in the contents.csv file.

```python
import os
import PyPDF2
from PyPDF2 import PdfFileWriter, PdfFileReader
import pandas as pd

# get the file names in the directory
directory = './pdf_docs'
entries = os.listdir(directory)

os.mkdir('./output')


df = pd.DataFrame(columns=['index', 'contents']) #empty dataframe 

count = 0

for entry in entries:
    print(entry)
    count += 1
        
    # create a PDF reader object
    pdfFileObj = open(directory + '/' + entry, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj, strict=False)
    

    # creating a page object
    pageObj = pdfReader.pages[0]

    # extracting text cotent from page
    content=(pageObj.extract_text())
    
    #extract and write to file
    with open(f'./output/file_{count}.txt', 'w', encoding="utf-8") as file:
        file.write(content)
         
    df_new_row = pd.DataFrame.from_records({ 'index': [count], 'contents': [content] })
    df = pd.concat([df, df_new_row])
    
    
    # closing the pdf file object
    pdfFileObj.close()
    
df.to_csv("./output/contents.csv", index=False, encoding='utf-8')

```
